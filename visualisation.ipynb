```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Visualising the data
This will provide a useful insight into my listening habits and music taste

```python
tracks = pd.read_csv('data/data.csv')
tracks
```

## Distribution of features

```python
featuresDf = tracks[['danceability','energy','loudness','tempo']]

fig, axs = plt.subplots(3, 3, figsize=(14, 14))
sns.histplot(tracks['danceability'], ax=axs[0,0])
sns.histplot(tracks['energy'], ax=axs[0,1])
sns.histplot(tracks['loudness'], ax=axs[0,2])
sns.histplot(tracks['tempo'], ax=axs[1,0])
sns.histplot(tracks['key'], ax=axs[1,1])
sns.histplot(tracks['speechiness'], ax=axs[1,2])
sns.histplot(tracks['acousticness'], ax=axs[2,0])
sns.histplot(tracks['instrumentalness'], ax=axs[2,1])
sns.histplot(tracks['valence'], ax=axs[2,2])
```
